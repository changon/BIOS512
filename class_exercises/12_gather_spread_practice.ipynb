Using the following data,

```r
library('tidyverse')

# From http://stackoverflow.com/questions/1181060
stocks = tibble(
  time = as.Date('2009-01-01') + 0:999,
  X = rnorm(1000, 0.35, 1),
  Y = rnorm(1000, 0, 2),
  Z = rnorm(1000, 2, 4)
)

stocks %>% print
```

Re-create the following figure:

(*Hints: use* `gather` *to re-shape the data.*)

<img src="figs/stock_gather.png" style="width: 500px;" />

```r
stocks.long = stocks %>% 
     as_tibble %>%
     gather(stock, price, -time)

stocks.long %>% head

p=ggplot(stocks.long, aes(x=stock, y=price), )
p=p+geom_point(position=position_jitter(h=0,w=.2), alpha=.5)

p
```
